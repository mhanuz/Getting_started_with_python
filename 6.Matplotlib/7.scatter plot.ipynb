```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
import numpy as np
```

```python
x=np.random.randint(1,20,10)
y=np.random.randint(5,15,10)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, s=100, c='red', edgecolor='black', alpha=.5)
```

```python
colors=np.random.randint(1,10,10)
```

```python
plt.scatter(x,y, s=100, c=colors, edgecolor='black', alpha=.5)
```

```python
plt.scatter(x,y, s=100, c=colors,cmap='Greens', edgecolor='black')
```

```python
plt.scatter(x,y, s=100, c=colors,cmap='Greens', edgecolor='black')
cbar=plt.colorbar()
cbar.set_label('Satisfaction')
```
