```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
slices=[60,40]
```

```python
plt.pie(slices)
plt.show()
```

```python
slices=[30,20,40,10]
labels=['thirty','twenty','forty','ten']
plt.pie(slices, labels=labels)
plt.legend()
plt.show()
```

```python
#wedge property
slices=[30,20,40,10]
labels=['thirty','twenty','forty','ten']
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})
plt.legend()
plt.show()
```

```python
#wedge property
slices=[30,20,40,10]
labels=['thirty','twenty','forty','ten']
colors=['red','green','blue','pink']
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'}, colors=colors)

plt.show()
```

```python
slices=[30,20,40,10]
labels=['thirty','twenty','forty','ten']
colors=['red','green','blue','pink']
explode=[0,.1,0,0]
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'}, colors=colors,explode=explode)
plt.show()
```

```python
slices=[30,20,40,10]
labels=['thirty','twenty','forty','ten']
colors=['red','green','blue','pink']
explode=[0,.1,0,0]
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'}, colors=colors,explode=explode, shadow=True)
plt.show()
```

```python
slices=[30,20,40,10]
labels=['thirty','twenty','forty','ten']
colors=['red','green','blue','pink']
explode=[0,.1,0,0]
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'}, colors=colors,
        explode=explode, shadow=True, startangle=90)
plt.show()
```

```python
slices=[30,20,40,10]
labels=['thirty','twenty','forty','ten']
colors=['red','green','blue','pink']
explode=[0,.1,0,0]
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'}, colors=colors,
        explode=explode, shadow=True, startangle=90, autopct='%1.1f%%')
plt.show()
```
