```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
df=pd.read_csv('developer.csv')
```

```python
df.head()
```

```python
ages=df['Age']
dev_salaries=df['All_Devs']
py_salaries=df['Python']
js_salaries=df['JavaScript']
```

```python
plt.plot(ages,dev_salaries, color='red', linestyle='--', label='salary')
plt.plot(ages,py_salaries, label='py salary')
plt.legend()
plt.show()
```

```python
plt.plot(ages,dev_salaries, color='red', linestyle='--', label='salary')
plt.plot(ages,py_salaries, label='py salary')
plt.fill_between(ages,py_salaries)
plt.legend()
plt.show()
```

```python
plt.plot(ages,dev_salaries, color='red', linestyle='--', label='salary')
plt.plot(ages,py_salaries, label='py salary')
overall_medial=57278
plt.fill_between(ages,py_salaries,overall_medial,alpha=.25)
plt.legend()
plt.show()
```

```python
plt.plot(ages,dev_salaries, color='red', linestyle='--', label='salary')
plt.plot(ages,py_salaries, label='py salary')
overall_medial=57278
plt.fill_between(ages,py_salaries,overall_medial,
                 where=(py_salaries>overall_medial), interpolate=True, alpha=.25)
plt.legend()
plt.show()
```

```python
plt.plot(ages,dev_salaries, color='red', linestyle='--', label='salary')
plt.plot(ages,py_salaries, label='py salary')
overall_medial=57278
plt.fill_between(ages,py_salaries,dev_salaries,
                 where=(py_salaries>dev_salaries), interpolate=True, alpha=.25)

plt.fill_between(ages,py_salaries,dev_salaries,
                 where=(py_salaries<=dev_salaries), interpolate=True, alpha=.25)
plt.legend()
plt.show()
```
