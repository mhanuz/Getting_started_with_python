```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
ages=[19,23,22,21,35,55,45,67,77,68,98,65]

plt.hist(ages, bins=10, edgecolor='black')
plt.title('Distrubution of ages')
plt.xlabel('ages')
plt.ylabel('salary')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
ages=[19,23,22,21,35,55,45,67,77,68,98,65]
bins=[10,20,30,40,50,60]
plt.hist(ages, bins=bins, edgecolor='black')
plt.title('Distrubution of ages')
plt.xlabel('ages')
plt.ylabel('salary')
plt.show()
```
